```python
#Some additional fucntions

#solve() , method used to solve linear equations using matrix as Ax = b, where A is a square matrix and b could be matrix or vector(1d array)
#consider an exmaple

'''
 6x + 4y = 18
 36x + 32y = 148

 now we will solve our equation using solve()



'''
A = np.array([[6,4],[36,32]])
B = np.array([18,148])

print(f"our A is \n {A} ")
print(f"B is \n {B}")
print(f'Using solve our x and y values are " {np.linalg.solve(A,B)}')

#But what if our matrix is not squeare ? Then we will use another function to solve our equation
# we will use lstsq, least square approach to solve the equation
# consider another example

'''
 6x + 4y = 18
 36x + 32y = 148
 5x + 24y = 98

 now we will solve our equation using solve()



'''
a = np.array([[6,4],[36,32],[5,24]])
b = np.array([18,148,98])

print(f"our A is \n {a} ")
print(f"B is \n {b}")
#print(f'Using least square our x and y values are " {np.linalg.lstsq(a,b)}')
x ,residuals,rank,s = np.linalg.lstsq(a,b,rcond = None)
print("Least squares solution for x and y:", x)
print("Residuals:", residuals)
print("Rank of matrix A:", rank)
print("Singular values of A:", s)
```

```python
#some other functions like qr,svd and eig

#QR decomposition. you can calculate matrix qr decomposition usign qr function
# like a matrix a = qr ,where a should be square matrix

'''
a =qr
q is an orthogonal (or unitary) matrix (its columns are orthonormal, meaning qTq = I)
r is an upper triangular matrix (all elements below the main diagonal are zero).
'''

q, r = np.linalg.qr(a)
print(f'using solve q is \n{q} \n and r is \n {q} \n ')


#To calculate single valued decomposition we will use svd()
#consider the bellow exmaple

'''
Singular Value Decomposition (SVD) is a more general matrix decomposition that applies to any m*n matrix (not necessarily square). It factors matrix
A into three matrices: A=UΣV^T

where ,
  U is an m*m orthogonal matrix (its columns are called the left singular vectors of A),
  Σ is an m*n diagonal matrix whose diagonal elements are the singular values of A,
  V^T is an n*n orthogonal matrix (its columns are the right singular vectors of A).

'''

U,sigma,V = np.linalg.svd(a)
print(f"using svd , U is \n {U} \n  \n sigma is {sigma} \n and V^T is \n {V}")


#To get eigenvectors and eigenvalues we can use eig() function

'''
Eigendecomposition is a factorization of a square matrix into its eigenvalues and eigenvectors. It is a fundamental concept in linear algebra and is useful in various fields such as data science, machine learning, and quantum physics.

For a given square matrix A, if there exists a vector v (called an eigenvector) and a scalar λ (called an eigenvalue) such that:

A v = λ v
'''

eigenvalues, eigenvectors = np.linalg.eig(A)
print(f"Using eigenvalues  and eigenvectors for our square matrix \n {A} \n the eigenvalue is {eigenvalues} and the eigenvector is \n {eigenvectors} ")
```

```python
#Handling Missing Data With Pandas
#we could delete missing values or NaN
#we can impute missing values by filling value instead of missing value or NaN
#let's get started

#drop missning values
data_with_nan = pd.DataFrame([[1., 6.5, 3.], [1., np.nan, np.nan],
                    [np.nan, np.nan, np.nan], [np.nan, 6.5, 3.]])
data_with_nan

#checking with isnull
data_with_nan.isnull()

#either we could do
data_with_nan[data_with_nan.notnull()]

#we can drop rows with nan values using dropna()
#data_with_nan.dropna()
data_with_nan.dropna() #we can pass 'how' argument to tell how to delete row or column. If we pass 'all' ,only the rows or columns contain all values as nan will be dropped
#pass how = 'all' and check

#we can drop column with nan using axis = 1
data_with_nan.dropna(axis = 1)

#we can also pass an argument called 'thresh'
data_with_nan.dropna(thresh = 2) #rows or columns(based on axis) which has more than or equal number not nan values will be returned
#check changing axis

```

```python
#Filling Missing values
#Dropping is not a optimal solution always
#We need to know filling missing data

#we can use fillna() methods to filling missing or nan values

#Either we can replace all nan with a constan like,
data_with_nan.fillna(0) #replacing nan with 0

#Or we can pass a dict like specifiying different value for each column
replace_val = {0:3,1:1,2:100} #dict , which contain a specified replacement value for nan for each diffrent column
data_with_nan.fillna(replace_val)

#fillna return a new object . to modify orignial dataframe we should pass 'inplace' as True
#data_with_nan.fillna(replace_val,inplace = True)

#Also we can use interpolation method as we have used in reindexing
data_with_nan.fillna(method ='ffill') #filling nan values with its forward index's value for the same column . consider [1,1] postion's nan will be replaced with [0,1] postion's value which is 6.5
#method = 'bfill' will fill the backward index's data . like [1,1] value will be replace with [2,1]'s value

#Also we can set limit , how many consecutive periods will be fill . Like limit = 1 only consecutive postion
data_with_nan.fillna(method = 'ffill' , limit = 1)

#we can do this fillna() in column wise too. like replace a nan value with its neighboring columns . depends on ffill or bfill
data_with_nan.fillna(method = 'ffill',limit = 2, axis = 1) #replacing [1,1]'s nan value with [1,0]'s value. do you get the difference with axis = 0?

#Even we can pass any aggr fucntion like mean or sum
data_with_nan.fillna(data_with_nan.mean()) #replacing every nan value with mean of each column of the dataframe

```

```python
#merging dataframe
#we can perform join operation like sql using merge() function
#Even using both column not necessarily the same name . I mean even if the names are different , we can handle

#consider two dataframe with a common column first
mer_df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
mer_df2 = pd.DataFrame({'key': ['a', 'b', 'd'],'data2': range(3)})

#as the both have key column , we can perform a merge on key
mer_df1.merge(mer_df2,on='key',how ='outer') #outer join with two dataframe

#Now if they do not have same column , still we can perform merge
mer_df3 = pd.DataFrame({'k': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
mer_df4 = pd.DataFrame({'key': ['a', 'b', 'd'],'data2': range(3)})

mer_df3.merge(mer_df4,left_on='k',right_on ='key',how = 'outer') #we will specify columns using left_on and right_on for left and right df respectively

'''
Param of how will be ,

'inner' : Use only the key combinations observed in both tables
'left' : Use all key combinations found in the left table
'right': Use all key combinations found in the right table
'output': Use all key combinations observed in both tables together
'''

#we can even merge on indexes or colunm with index
#we can make right dataframe's index to key using right_index = True and for using left dataframe's index to key, left_index = True

left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],'value': range(6)})
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])

#using left1's key and right1's index
left1.merge(right1,left_on ='key',right_index = True, how  ='inner')

#your task is to join using both dataframe's index
#code here






#Or let me know . I will provide later notebook
#if you want to join on index only , you also can use join() function

left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],index=['a', 'c', 'e'],columns=['Ohio', 'Nevada'])
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],index=['b', 'c', 'd', 'e'],columns=['Missouri', 'Alabama'])

left2.join(right2,how = 'inner')

#we can join() even columns using on
#even you can join() a list of dataframe using a list of dataframe like .join([dataframe_one,dataframe_two])
```

```python
#Groupby in Pandas

import pandas as pd

# Sample DataFrame
data = {
    'Salesperson': ['Alice', 'Bob', 'Alice', 'Bob', 'Alice', 'Bob', 'Alice', 'Bob'],
    'Region': ['East', 'West', 'East', 'West', 'East', 'West', 'East', 'West'],
    'Quarter': ['Q1', 'Q1', 'Q2', 'Q2', 'Q3', 'Q3', 'Q4', 'Q4'],
    'Sales': [200, 180, 220, 210, 230, 190, 240, 200]
}

# Create DataFrame and set multi-index with Salesperson and Region
df = pd.DataFrame(data)
df.set_index(['Salesperson', 'Region'], inplace=True)  # Setting a multi-index for later grouping
print("Initial DataFrame with multi-index:\n", df, "\n")

# Example 1: Group by 'Quarter' and sum the sales for each quarter
grouped_by_quarter = df.groupby('Quarter').sum()
print("Total sales for each quarter:\n", grouped_by_quarter, "\n")

# Example 2: Group by 'Salesperson' and 'Region' using multi-index and sum sales
grouped_by_salesperson_region = df.groupby(level=['Salesperson', 'Region']).sum()
print("Total sales by Salesperson and Region:\n", grouped_by_salesperson_region, "\n")

# Example 3: Group by 'Quarter' and 'Region' and calculate mean sales for each group
grouped_by_quarter_region = df.groupby(['Quarter', 'Region']).agg({'Sales': 'mean'})
print("Mean sales by Quarter and Region:\n", grouped_by_quarter_region, "\n")

# Example 4: Group by 'Quarter' with as_index=False to avoid setting 'Quarter' as the index in the result
grouped_no_index = df.groupby(['Quarter'], as_index=False).sum()
print("Total sales by Quarter without setting 'Quarter' as index:\n", grouped_no_index, "\n")

# Example 5: Group by 'Salesperson' (using index level) and sum sales
grouped_by_salesperson = df.groupby(level='Salesperson').sum()
print("Total sales by Salesperson:\n", grouped_by_salesperson)
```
